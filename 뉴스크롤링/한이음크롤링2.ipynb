```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html=urlopen("http://news.hankyung.com/article/2017071626111")
soup=BeautifulSoup(html,'html.parser')
s=soup.find_all('div', {'id': "newsView"})
s
```

```python
s=str(s)
while True:
    a=s.find('<')
    if a==-1:
        break
    b=s.find('>')+1
    s=s[:a]+s[b:]
print(s)
```

```python
s.replace("\n","").replace("\t","").replace("\r","")
```
